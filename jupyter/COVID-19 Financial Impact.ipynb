# Stock Indices

```python
# Build the endpoint URL for DJI, GSPC, IXIC, VIX
# Time : one day interval
target_url = (f'https://api.twelvedata.com/time_series?symbol=DJI,GSPC,IXIC,VIX&start_date=2020-01-01&interval=1day&apikey={key}')
```

```python
# Run a request to endpoint and convert result to json
stock_data = requests.get(target_url).json()
```

```python
# Print the json (pretty printed)
print(json.dumps(stock_data, indent=4, sort_keys=True))
```

```python
stocks = list(stock_data.keys())
```

```python
mylist = []
for stock in stocks:
    values = stock_data[stock]["values"]
    for value in values:
        value["name"] = stock
        mylist.append (value)
```

# COVID-19 by country - US

```python
# Build the endpoint URL for COVID-19 by country
# Returns confirmed type for a country from the first recorded case. 
# Source : https://covid19api.com/
target_url = ('https://api.covid19api.com/total/dayone/country/us/status/confirmed')
```

```python
# Run a request to endpoint and convert result to json
US_data = requests.get(target_url).json()
```

```python
# Print the json (pretty printed)
print(json.dumps(US_data, indent=4, sort_keys=True))
```

```python
US_df = pd.DataFrame (US_data).reset_index()
US_df = US_df.rename(columns={"index":"id"})
```

```python
#convert to csv file
US_df.to_csv("US_Covid19.csv",index=False)
US_df.head()
```
